```python
import cv2
import matplotlib.pyplot as plt

# 
img = cv2.imread('Lounge1.jpg', cv2.IMREAD_GRAYSCALE)

# Scharr
scharr_x = cv2.Scharr(img, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(img, cv2.CV_64F, 0, 1)
scharr = cv2.magnitude(scharr_x, scharr_y)

# 8bit
scharr_uint8 = cv2.normalize(scharr, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
cv2.imwrite('Lounge1_scharr.jpg', scharr_uint8)

# 
plt.figure(figsize=(10,5))

# 
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

# Scharr
plt.subplot(1, 2, 2)
plt.imshow(scharr, cmap='gray')
plt.title('Scharr Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# 
img_color = cv2.imread('Lounge1.jpg')

# 
img_gray = cv2.imread('Lounge1.jpg', cv2.IMREAD_GRAYSCALE)

# Scharr
scharr_x = cv2.Scharr(img_gray, cv2.CV_64F, 1, 0)
scharr_y = cv2.Scharr(img_gray, cv2.CV_64F, 0, 1)
scharr = cv2.magnitude(scharr_x, scharr_y)

# 8bit
scharr_uint8 = cv2.normalize(scharr, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
cv2.imwrite('Lounge1_scharr.jpg', scharr_uint8)

# 
plt.figure(figsize=(15,5))

# 
plt.subplot(1, 3, 1)
img_color_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)  # BGRRGB
plt.imshow(img_color_rgb)
plt.title('Original Color Image')
plt.axis('off')

# 
plt.subplot(1, 3, 2)
plt.imshow(img_gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

# Scharr
plt.subplot(1, 3, 3)
plt.imshow(scharr, cmap='gray')
plt.title('Scharr Edge Detection')
plt.axis('off')

plt.tight_layout()
plt.show()
```
